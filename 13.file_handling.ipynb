# File handling

```python
# Read a whole file at once
with open('example.txt','r') as file:
    content =file.read()
    print(content)
```

```python
# Read file line by line
with open('example.txt','r') as file:
    for lines in file:
        print(lines.strip())  # strip() to remove extra newlines
```

```python
# Write file as overwrite
with open('example.txt', 'w') as file:
    file.write("This is an new line.\n")
```

```python
# Write to a file
with open('example.txt', 'a') as file: # append mode
    file.write("\nadd a new line.\n")
```

```python
# write multiple lines
lines = ["First line.\n", "Second line.\n", "Third line.\n"]
with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
# write binaary file
data= b'\x00\x01\x02\x03\x04\x05'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
# read binary file
with open('example.bin','rb') as file:
    print(file.read())
```

```python
# write then read a file (W+)
with open('example.txt','w+') as file:
    file.write("Writing and reading in same file.\n")
    file.write('hello')
    file.seek(0) # cursor index  
    print(file.read())
```

```python
# finding current working directory
import os
current_directory = os.getcwd()
print("Current Working Directory:", current_directory)
```

```python
# create a new directory
new_directory = 'my_new_directory'
import os
os.mkdir(new_directory)
```

```python
# list files & directories in current directory
items = os.listdir('.') # . represents current directory
print(items)
```

```python
# joining paths
dir_name = 'my_new_directory'
file_name = 'example.txt'
full_path = os.path.join(dir_name, file_name)
print("Full Path:", full_path)
```

```python
# path existence check
path_to_check = 'example1.txt'
print(os.path.exists(path_to_check))
```

```python
# checking a path is file or directory
path_to_check = 'my_new_directory'
if os.path.isfile(path_to_check):
    print(f"{path_to_check} is a file.")
elif os.path.isdir(path_to_check):
    print(f"{path_to_check} is a directory.")
else:
    print(f"{path_to_check} does not exist.")
```

```python
# getting absulute path
relative_path = 'example.txt'
absolute_path = os.path.abspath(relative_path)
print("Absolute Path:", absolute_path)
```
