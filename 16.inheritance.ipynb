# Inheritance

```python
# Single Inheritance 
# parent class
class Car:
    # constructor
    def __init__(self,window,doors,engine_type):
        self.window=window
        self.doors=doors
        self.engine_type=engine_type
    def engine(self):
        print(f"is a {self.engine_type} car ")
```

```python
car1=Car(4,4,'Petrol')
car1.engine()
```

```python
# Child class
class Tesla(Car):
    # Constructor
    def __init__(self,window,doors,engine_type,is_selfdrive):
        super().__init__(window,doors,engine_type)
        self.is_selfdrive=is_selfdrive
    # method 
    def selfDrive(self):
        print(f"Tesla support selfdrive: {self.is_selfdrive}")

```

```python
tesla1=Tesla(2,2,'Electic',True)
tesla1.selfDrive()
```

```python
tesla1.engine()
```

```python
# Multiple Inheritance
# Base class
class Animal:
    def __init__(self,name):
        self.name=name
    def bark(self):
        print(f"{self.name} can bark")

# Another Base Class
class Pet:
    def __init__(self,age):
        self.age=age
    def older(self):
        print(f"{self.age} older")

# derived class 
class Dog(Animal,Pet):
    def __init__(self,name,age):
        Animal.__init__(self,name)
        Pet.__init__(self,age)
    def speak(self):
        print(f"{self.name} bark ghew")
```

```python
# create an object
dog1=Dog("tommy",5)
```

```python
dog1.older()
```

```python
dog1.bark()
```

```python
dog1.speak()
```
