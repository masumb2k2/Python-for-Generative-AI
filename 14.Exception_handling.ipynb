# Exception Handling

```python
# try except block
try:
    a=b
except:
    print('Varibale is not defined')
```

```python
try:
    a=c
except NameError as ex:
    print(ex)
```

```python
try:
    result=10/1
    d=e
except ZeroDivisionError as ex:
    print(ex)
except Exception as ex:
    print(ex)
```

```python
# try, Except and else block 
try:
    num=int(input("enter Any number"))
    result=12/num
except ValueError as ex:
    print("this input is not valid")
except ZeroDivisionError as ex:
    print(ex)
except Exception as ex:
    print(f"Error type {ex}")
else:
    print(result)
```

```python
# try, Except and else  and finally block 
try:
    num=int(input("enter Any number"))
    result=12/num
except ValueError as ex:
    print("this input is not valid")
except ZeroDivisionError as ex:
    print(ex)
except Exception as ex:
    print(f"Error type {ex}")
else:
    print(result)
finally:
    print('code execute')
```

```python
# file Handling ith Exception
try:
    file=open('random.txt','r')
    content=file.read()
    print(content)
except FileNotFoundError as ex:
    print(ex)
except Exception as ex:
    print(ex)
finally:
    if 'file' in locals() and not file.closed():
        file.close()
        print('file closed')
```
