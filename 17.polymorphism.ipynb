# Polymorphism

```python
# Method Overriding

# Base Case
class Animal:
    def speak(self):
        return 'this is the sound of animal'
# Derive class 1
class Dog(Animal):
    def speak(self):
        return 'ghew'
# Dreive class 2
class Cat(Animal):
    def speak(self):
        return 'Meaw'

# fuction to implement polymorphism
def animal_speak(animal):
    print(animal.speak())

# object 
tommy=Dog()
dora=Cat()

print(tommy.speak())

animal_speak(dora)

```

```python
# ABstracion base  
from abc import ABC,abstractmethod #ABC: abstracion base class

# Base Class
class Vechile(ABC):
    @abstractmethod
    def engine_type(self):
        pass

# Derive class 1
class Car(Vechile):
    def engine_type(self):
        return "Engine type is Petrol"
# another Drived class
class Track(Vechile):
    def engine_type(self):
        return 'Engine type is disel'

#Function for polymorphism
def fueling(random):
    print(random.engine_type())

# object Creation
car1=Car()
track1=Track()

fueling(car1)
fueling(track1)
```
