```python
# even odd find 
num=12
if num%2==0:
    print('even')
else:
    print('odd')
```

```python
def even_or_odd(num):
    '''this function is used to find even or odd'''
    if num%2==0:
        print('even')
    else:
        print('odd')
```

```python
even_or_odd(13)
```

```python
even_or_odd(66)
```

```python
# positional arguments 

def print_value(*baut):
    for value in baut:
        print(value)
```

```python
print_value(10,20,30,40,50,56,66,'masum')
```

```python
# keyword arguments
def print_details(**inf0):
    for key,value in inf0.items():
        print(f"{key} : {value}")
```

```python
print_details(name='Masum',age=24,city='Dhaka')
```
