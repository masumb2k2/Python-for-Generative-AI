```python
student={
    "name":"Masum",
    "age":24,
    'id': 1
}
```

```python
print(student['age']) # accessing value using key
```

```python
# moedifying dictionary values
student
```

```python
student['id']=100 # updating value
student
```

```python
student['grade']='A+' # adding new key-value pair
student
```

```python
del student['age'] # deleting a key-value pair
student
```

```python
# dictionary methods
keys=student.keys()
print(keys)
value=student.values()
print(value)
items=student.items()
print(items)
```

```python
# iteration over dictionaries
for key in student.keys():
    print(key)
```

```python
for x in student.values():
    print(x)
```

```python
for key,value in student.items():
    print(key,value)
```

```python
# nested dictionaries
students={
    'student1':{'name':'Masum',
        'age':24
    },
    'student2':{
        'name':'John',
        'age':22
    }
}
students
```

```python
# dictionary comprehension
squares={x:x**2 for x in range(10)} 
squares
```

```python
squares={x:x**2 for x in range(10) if x%2==0}
squares 
```
