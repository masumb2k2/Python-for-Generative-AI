# File handling

```python
# Read a whole file 
with open('example.txt','r') as file:
    print(file.read())
```

```python
#read line by line
with open('example.txt','r') as file:
    for lines in file:
        print(lines.strip())
```

```python
# write a whole file 
with open('example.txt','w')as file:
    file.write('i write this file')
```

```python
# Write wihtout overwrite
with open('example.txt','a') as file:
    file.write('\nthis is a new line')
```

```python
# write multiple line
lines=['first line','second line','third line']
with open('example.txt','a') as file:
    file.writelines(lines)
```
