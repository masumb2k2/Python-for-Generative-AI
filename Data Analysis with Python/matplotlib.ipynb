It is the most widely used tool for data visualization in the Python ecosystem and serves as the foundation for many other high-level plotting libraries, such as Seaborn and Pandas.

```python
import matplotlib.pyplot as plt
```

```python
# Single Plot
x=[1,2,3,4,5]
y=[23,4,12,8,2]

# create a line 
plt.plot(x,y)
plt.xlabel("this is X axis")
plt.ylabel("Ami Y axis")
plt.title("Ami chobi")
plt.show()
```

```python
# Customize plot
x=[1,2,3,4,5]
y=[23,4,12,8,2]

plt.plot(x,y,color='purple',marker='o',linestyle='--',linewidth=3,markersize=8)
plt.xlabel("this is X axis")
plt.ylabel("Ami Y axis")
plt.title("Ami plot")
plt.show()
```

```python
# multiple plots in one figure
x=[1,2,3,4,5]
y1=[23,4,12,8,2]   
y2=[10,20,25,30,40]
y3=[5,15,22,18,28]
y4=[7,17,27,37,47]
plt.figure(figsize=(10,5))

plt.subplot(2,2,1) # row 2  column 2  plot 1
plt.plot(x,y1,color='purple',marker='o',linestyle='--',linewidth=3,markersize=8)
plt.title("First Plot")

plt.subplot(2,2,2)
plt.plot(x,y2,color='green',marker='o',linestyle='--',linewidth=3,markersize=8)
plt.title("Second Plot")

plt.subplot(2,2,4)
plt.plot(x,y3,color='red',marker='o',linestyle='--',linewidth=3,markersize=8)
plt.title("Third Plot")

plt.subplot(2,2,3)
plt.plot(x,y4,color='blue',marker='o',linestyle='--',linewidth=3,markersize=8)
plt.title("Fourth Plot")
```

```python
# multiple plots in one figure
x=[1,2,3,4,5]
y1=[23,4,12,8,2]   
y2=[10,20,25,30,40]

plt.figure(figsize=(10,5))

plt.subplot(1,2,1) # row 2  column 2  plot 1
plt.plot(x,y1,color='purple',marker='o',linestyle='--',linewidth=3,markersize=8)
plt.title("First Plot")

plt.subplot(1,2,2)
plt.plot(x,y2,color='green',marker='o',linestyle='--',linewidth=3,markersize=8)
plt.title("Second Plot")

```

```python
# Bar Plot
categories=['Mango','Orange','Bananna','Pine-apple']
amount=[20,14,19,3]

# Crate a bar plot
plt.bar(categories,amount,color='teal')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('BAr plot')
```

A histogram is a type of graphical representation used in statistics to show the distribution of numerical data

```python
data=[10,20,20,30,30,30,40,40,40,40]

# Create a histrogram
plt.hist(data,bins=4,color='orange',edgecolor='black')
```

```python
# A scatter plot is a graph that uses dots to show the relationship between two different numerical variables

# same date
x=[1,2,3,4,5]
y=[2,3,6,8,9]

plt.scatter(x,y,color='teal',marker='X')
```

```python
# pi chart 
lables=['Mango','Orange','Bananna','Pine-apple']
sizes=[30,10,40,20]
colors=['gold','orange','lightcoral','skyblue']
explode=(0.2,0,0,0)

# CREATE THE BAR CHART
plt.pie(sizes,labels=lables,explode=explode, colors=colors,autopct='%1.1f%%',shadow=True)
```
