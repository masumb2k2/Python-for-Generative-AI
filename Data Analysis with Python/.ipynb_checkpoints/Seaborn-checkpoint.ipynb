Seaborn is a popular Python library for statistical data visualization. It is built on top of the core Matplotlib library and provides a high-level interface for creating attractive and informative statistical graphics with less code. Seaborn works seamlessly with data structures from the pandas library, making it a favorite tool for data scientists and analysts.

```python
import seaborn as sns
```

```python
# Basic plotting with seaborn 
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
# Creata a scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('Scatter Plot')
plt.xlabel("x axis")
plt.ylabel('Y axis')
plt.show()
```

```python
# Line plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('Lineplot')
plt.show()
```

```python
tips.head()
```

```python
# categorical plot 
# bar plot 
sns.barplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
# Box plot 
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# violin plot 
sns.violinplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
# Histrogram  Kernel Density Estimate (KDE
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
# KDE plot 
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
tips.head()
```

```python
sns.pairplot(tips)
```

```python
# heatmeap 
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
