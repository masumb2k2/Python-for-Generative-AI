Series: A one-dimensional labeled array capable of holding any data type, similar to a single column in a spreadsheet.

```python
# series from list
data=[1,2,3,4,5]
srs=pd.Series(data)
print(srs)
print(type(srs))
```

```python
# Create a series from dictionaries
data={
    'a':1,
    'b':2,
    'c':3
}
d_srs=pd.Series(data)
d_srs
```

```python
data=[1,2,3,4,5]
ind=['a','b','c','d','e']
i_srs=pd.Series(data,index=ind)
i_srs
```

DataFrame: A two-dimensional, size-mutable, tabular data structure with labeled axes (rows and columns), resembling a spreadsheet or SQL table. The DataFrame is the most commonly used pandas object.  

```python
# create a datafram from dictionary of lists
data={
    'name':['masum','sabrina','puja','oishe'],
    'age':[23,24,26,30],
    'city':['dhaka','khulna','faridpur','natore']
}

dframe=pd.DataFrame(data)
print(dframe)
print(type(dframe))
```

```python
# create a datafram from list of dictionary 
data=[
    {'name':'Masum','age':23,'city':'dhaka'},
    {'name':'Sabrina','age':23,'city':'dhaka'},
    {'name':'puja','age':23,'city':'dhaka'},
    {'name':'oishe','age':23,'city':'dhaka'}
]
ldframe=pd.DataFrame(data)
ldframe
```
