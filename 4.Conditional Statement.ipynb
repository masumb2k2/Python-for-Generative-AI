```python
# if Statement 
age = int(input("Enter your age: "))

if age >= 18:
    print("You are an adult.")
```

```python
# else  Statement 
age = int(input("Enter your age: "))

if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
```

```python
age=12
if age<13:
    print("You are a child.")
if age<18:
    print("You are a teenager.")  
```

```python
age=12
if age<13:
    print("You are a child.")
elif age<18:
    print("You are a teenager.")  
```

```python
# elif Statement
age=12
if age<13:
    print("You are a child.")
elif age<18:
    print("You are a teenager.")   
else:
    print("You are an adult.")
```

```python
# Nested conditional Statements
 
age =20
salary = 40000

if age <30:
    if salary >40000:
        print("eligible for marriage")
    else:
        print("not eligible for marriage")
else:
    print("not eligible for marriage")
```
