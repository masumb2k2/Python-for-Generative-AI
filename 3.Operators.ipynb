# Operators

```python
num1=10
num2=20
```

```python
# Ariothmetic Operators: +, -, *, /, %, //, **
sum=num1+num2
print("Sum:",sum)
```

```python
sub=num2-num1
print("Subtraction:",sub)
```

```python
mul=num1*num2
print("Multiplication:",mul)
```

```python
div=num2/num1
print("Division:",div)
```

```python
num3=56
num4=10
div2=num3/num4
print("Division 2:",div2)
```

```python
flrdiv=num3//num4 
print("Floor Division:",flrdiv)
```

```python
modulas=num3 % num4
print("Modulas:",modulas)
```

```python
exponential=2**6
print("Exponential:",exponential)
```

```python
# Assignment Operators: =, +=, -=, *=, /=, %=, //=, **=
a=5
print("Initial a:",a)

a=a+3


a+=3    # a=a+3
print("After a+=3:",a)
```

```python
a=5
b=8
a+=b # a=a+b
print("After a+=b:",a)
```
