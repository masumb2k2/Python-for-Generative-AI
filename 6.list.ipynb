```python
# iterating through a list
fruits = ['apple', 'banana', 'cherry', 'date', 'elderberry']
for fruit in fruits:
    print(fruit)
```

```python
# iterating through a list with index
for index,fruit in enumerate(fruits):
    print(index,fruit)
```

```python
# list comprehension
lst= []
for i in range(10):
    lst.append(i**2)
print(lst)
```

```python
lst1=[i**2 for i in range(10)]
lst1
```

```python
 #Basic sytex of list comprehension:
 # [expression for item in iterable]
# with optional condition:
 # [expression for item in iterable if condition]
```

```python
lst=[]
for i in range(10):
    if i%2==0:
        lst.append(i)
print(lst)
```

```python
even_number=[i**2 for i in range(10) if i%2==0]
print(even_number)
```

```python
# NESTED LIST COMPREHENSION
list1=[1,2,3]
list2=['a','b','c']

pair=[[i,j] for i in list1 for j in list2]
print(pair)
```
