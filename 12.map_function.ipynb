# Map function

```python
def squares(num):
    return num*num 
```

```python
squares(12)
```

```python
# map Syntax
list(map(function, iterable))
```

```python
numbers=[1,2,14,5,6,8]
list(map(squares,numbers))
```

```python
numbers=[1,2,14,5,6,8]
list(map(lambda x:x*x,numbers))
```

```python
# map with multiple iterable 
numbers1=[1,2,3,4,5]
numbers2=[4,5,6,7,8]
list(map(lambda num1,num2:num1*num2,numbers1,numbers2))
```
