```python
# While Loop
## syntax : initialization
## while condition: 
##     statement
##    increment/decrement


i=1
while i<=5:  # false  
    print(i)
    i+=1 # i=i+1
```

```python
# break 
i=1
while i<=10:
    print(i)
    if i==5:  # false
        break
    i+=1
```

```python
# continue
i=1
while i<8:
    i+=1
    if i==5:  # false
        continue
    print(i)
 
```

```python
# pass 
i=1
while i<=5:
    if i==3:
        print("This is pass block")
        pass
    else:
        print(i)
    i+=1      
```
